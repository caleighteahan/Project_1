```python
Med_sale_path = ("data/California_yearly_med_sale_price_value.csv")
med_sale_df = pd.read_csv(Med_sale_path)
new_med_sale_df = med_sale_df.drop(med_sale_df.columns[48:52],axis=1)
new_med_sale_df = med_sale_df.drop(med_sale_df.columns[0],axis=1)
new_med_sale_df.head()
```

```python
clean_df = new_med_sale_df.iloc[:, ::3]
clean_df
```

```python
months = clean_df.loc[0,:]
values = clean_df.loc[1,:]





x_axis = np.arange(len(months))
y_axis = [543, 586, 566, 575, 568, 566, 634, 652, 664, 757, 745, 753, 755, 835, 757, 733, 704]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
regress_values = slope * x_axis + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(months, values, marker='o', label = "California" )
plt.plot(x_axis, regress_values, "r-")
# plt.annotate(line_eq, (5.8, 0.8), color = "red")
# plt.figure(figsize=(12,8))
plt.xlabel("Months")
plt.ylabel("Median Price")
plt.title("Median Sales Price")
plt.xticks(months, rotation='vertical')
plt.legend()
plt.show()
```
